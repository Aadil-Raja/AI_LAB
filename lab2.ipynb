```python
#Pubg practice class task

import random

class Environment:
    def __init__(self):
        self.state = self._get_random_state()

    def _get_random_state(self):

        return random.choice(['Enemy Nearby',
                              'Not in SafeZone',
                              'Health 0',  'Low Ammo',
                              'Enemy Attacking',
            'In Vehicle',
            'Found Loot'])

    def get_percept(self):
        return self.state

    def update_state(self):

        self.state = self._get_random_state()

class SimpleReflexAgent:
    def __init__(self):
        pass

    def act(self, percept):

        if percept == 'Enemy Nearby':
            return 'Kill the enemy'
        elif percept == 'Not in SafeZone':
            return 'Move to Zone'
        elif percept == 'Health 0':
            return 'Apply Med-kit'
        elif percept == 'Low Ammo':
            return 'Reload'
        elif percept == 'Enemy Attacking':
            return 'Take Cover'
        elif percept == 'In Vehicle':
            return 'Drive to Safety'
        elif percept == 'Found Loot':
            return 'Pick Up Loot'
        else:
            return 'Search for Enemies'

def run_agent(agent, environment, steps):
    for step in range(steps):
        percept = environment.get_percept()
        action = agent.act(percept)
        print(f"Step {step + 1}: Percept - {percept}, Action - {action}")

        environment.update_state()


agent = SimpleReflexAgent()
environment = Environment()


run_agent(agent, environment, 10)
```

```python
#LAB TASK QUESTION 2

import random

class System:
  def __init__(self):
    self.servers = []
    for i in range(5):
      self.servers.append(random.choice(["Underloaded", "Balanced", "Overloaded"]))
  def display_servers(self):
        x = 'A'
        for i in range(5):
            print(f"Server {x}: {self.servers[i]}")
            x = chr(ord(x) + 1)

class SimpleReflexAgent:
  def __init__(self):
    pass
  def balance_load(self, data_center):
        underloaded = []
        overloaded = []
        for i in range(5):
            if data_center.servers[i] == "Underloaded":
                underloaded.append(i)
            elif data_center.servers[i] == "Overloaded":
                overloaded.append(i)
        while overloaded and underloaded:
            o_index = overloaded.pop()
            u_index = underloaded.pop()
            data_center.servers[o_index] = "Balanced"
            data_center.servers[u_index] = "Balanced"
        print("\nLoad balancing completed. Updated server statuses:")
        data_center.display_servers()


data_center = System()
agent = SimpleReflexAgent()

print("Initial Server Load Status:")
data_center.display_servers()

print("\nBalancing the Load...")
agent.balance_load(data_center)
```

```python
#LAB TASK QUESTION 4

import random


class System:
  def __init__(self):
    self.components={}
    for component in "ABCDEFGHI":
       self.components[component] = random.choice(["Safe", "Low Risk Vulnerable", "High Risk Vulnerable"])
  def display(self):
    print("\nCurrent:")
    for component,status in self.components.items():
      print(f"Component {component} : {status}")

class UtilityBasedAgent:
  def __init__(self):
     self.utility = {"Safe": 10, "Low Risk Vulnerable": -5, "High Risk Vulnerable": -10}

  def calculate_utility(self, status):
    return self.utility[status]

  def scan_and_patch(self, system):
        print("Scanning system for vulnerabilities.")
        total_utility = 0
        for component, status in system.components.items():
            utility_score = self.utility[status]
            total_utility += utility_score
            if status == "Safe":
                print(f"\nComponent {component} is secure. Utility: {utility_score}")
            elif status == "Low Risk Vulnerable":
                print(f"\nWarning: Component {component} has a Low Risk Vulnerability! Utility: {utility_score},Fixing it")
                system.components[component] = "Safe"
                print(f"Component {component} has been patched and is now Safe.")
            elif status == "High Risk Vulnerable":
                print(f"\nComponent {component} : High Risk Vulnerability and requires a premiumservice to patch! Utility: {utility_score},Can not fix it ")
                print("Do u want to buy Premium yes/no")
                x=random.choice(["yes","no"])
                if x=='yes':
                   print("Buying premium Service to fix\n")
                   system.components[component] = "Safe"
                else:
                  print("not buying\n")

        print(f"\nTotal Utility Score: {total_utility}")

system=System()
system.display()
print("\n")
agent= UtilityBasedAgent()
agent.scan_and_patch(system)

system.display()
```

```python
#LAB TASK QUESTION 6

class Environment:
    def __init__(self):
        self.rooms = {
            'a': 'Safe', 'b': 'Safe', 'c': 'Fire',
            'd': 'Safe', 'e': 'Fire', 'f': 'Safe',
            'g': 'Safe', 'h': 'Safe', 'j': 'Fire'}

    def display_environment(self):
        print("\nCurrent Environment Status:")
        layout = [['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'j']]
        for row in layout:
            row_display = []
            for room in row:
                if self.rooms[room] == "Fire":
                    row_display.append("ðŸ”¥")
                else:
                    row_display.append(" ")
            print(row_display)
        print("-----------------")

class Robot:
    def __init__(self):
        self.current_position = 'a'
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']

    def scan(self, environment):
        print("\nStarting Firefighting Operations")

        for room in self.path:
            self.current_position = room
            print(f"Robot moved to Room {room}.")
            if environment.rooms[room] == "Fire":
                print(f"Fire detected in Room {room}! Extinguishing fire")
                environment.rooms[room] = "Safe"
                print(f"Room {room} is now safe.")
            else:
                print(f"Room {room} is already safe.")
            environment.display_environment()

        print("\nFirefighting complete. All rooms are now safe!")

environment = Environment()
robot = Robot()
environment.display_environment()
robot.scan(environment)


```
